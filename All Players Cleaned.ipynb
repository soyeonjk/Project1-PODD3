```python
# Import dependencies
import pandas as pd
```

```python
# Name of the CSV file and show path
file = "All_playerss.csv"
all_playerss_csv_path = "All_playerss.csv"
```

```python
all_players_clean = "All_playerss.csv"
all_players_clean = pd.read_csv(file)
all_players_clean
```

```python
# Import the data into a Pandas DataFrame
players_df = pd.read_csv(all_playerss_csv_path)
players_df
```

```python
# Indexing and eliminate rows past 2019
indexNames = all_players_clean[ all_players_clean["Year"] > 2019 ].index
# Delete rows from dataFrame
all_players_clean.drop(indexNames , inplace=True)
all_players_clean
```

```python
# Drop all players with less than 900 Minutes
indexNames = all_players_clean[ all_players_clean["MINS"] < 900 ].index
# Delete rows from dataFrame
all_players_clean.drop(indexNames , inplace=True)
all_players_clean
```

```python
# Remove and delete unecessary columns from the DataFrame and save the new DataFrame
all_players_cleaned = all_players_clean.drop(["POS", "GP", "GS", "A", "SHTS", "HomeGame", "RoadGame", "SC%", "HmA", "RdA", "FS", "OFF"], axis = 1)
all_players_cleaned.head(10)

all_players_cleaned.sort_values(by = ["Club", "G"], ascending = False)
```

```python
players_byclubs = players_df.sort_values(by = ['Player', 'Club'], ascending = False)
players_byclubs
```

```python
players_byclubs = players_df.sort_values(by = ['Player', 'Club', 'G', 'GWG','G/90min','SOG%'], ascending = False)
players_byclubs
```
